```python
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd
from pandas import Series, DataFrame
from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['SimHei'] 


xls_file = pd.ExcelFile('cp_figure4.xlsx') 
table = xls_file.parse('Sheet1')  


df = table.set_index('Date')       
df.tail()                        


fig = plt.figure(figsize=(8,4.4))

ax1 = fig.add_subplot(111)

df['VR'].plot(ax=ax1,style='orange',alpha=1,label='VR',linestyle='-',linewidth=1,marker='x')    
plt.xlabel('Date')

ax1.set_yticks(np.arange(0,450000,50000))

ax1.set_ylabel('Number of posts containing #VR')

plt.legend(loc=2)


plt.title('The corresponding positions of the VR headsets launch time')
 
plt.savefig('CPfig4.png', dpi=600, bbox_inches='tight') 
```
